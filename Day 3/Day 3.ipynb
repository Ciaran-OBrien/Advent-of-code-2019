```python
def plot_wire(data):
    data = data.split(',')
    x,y = 0,0
    inflection_points = [[0,0]]
    for command in data:
        match = re.match(r"([A-Z])(\d+)",command)
        if match[1] == 'R':
            x += int(match[2])
            inflection_points.append([x,y])
        if match[1] == 'U':
            y += int(match[2])
            inflection_points.append([x,y])
        if match[1] == 'D':
            y -= int(match[2])
            inflection_points.append([x,y])
        if match[1] == 'L':
            x -= int(match[2])
            inflection_points.append([x,y])

    return x,y,inflection_points
```

```python
x,y,points_0 = plot_wire(data[0])
```

```python
x,y,points_1 = plot_wire(data[1])
```

```python
def process_wire(instr_line):
    current_pos = [0, 0]
    for instr in instr_line.split(','):
        for _ in range(int(instr[1:])):
            current_pos[0 if instr[0] in ('L', 'R') else 1] += -1 if instr[0] in ('L', 'D') else 1
            yield tuple(current_pos)
```

```python
with open(os.getcwd()+'\day3.txt', 'r') as f:
    wires = [list(process_wire(line)) for line in f.readlines()]
intersections = set(wires[0]) & set(wires[1])
print(min(abs(x)+abs(y) for (x, y) in intersections)) #Part 1
```
